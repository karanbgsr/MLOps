```python
df_test = pd.DataFrame({'SKU':['A','A','A','A','A','A','A','A','A'],'Sales':[10,5,1,2,100,1,4,3,100]})
```

```python
def moving_avg(ma =3, df = df):
    for i in range(ma):
        df = df.assign(**{f"Lead_{i+1}" : df['Sales'].shift(i+1)})
    return df


```

```python
df_output = moving_avg(3,df =df_test )
display(df_output)

df_output['ma3'] = df_output[['Lead_1','Lead_2','Lead_3']].mean(axis=1,skipna=True)
display(df_output)
```
